```python
# pip install pillow
import numpy as np
import matplotlib.pyplot as plt
from PIL import *
```

```python

image =  Image.open("input/Unequalized_Hawkes_Bay_NZ.jpg")
image = image.convert('L')
image = ImageOps.equalize(image)
plt.plot(image)
plt.show()
image.save('equalized.jpg')
```

```python

image = Image.open("input/tsukuba_l.png")
# Convert the image to grayscale
gray_image = image.convert('L')
gray_image = ImageOps.equalize(gray_image)
plt.plot(gray_image)
plt.show()
# Save the grayscale image
gray_image.save("example_gray.jpg")
```
