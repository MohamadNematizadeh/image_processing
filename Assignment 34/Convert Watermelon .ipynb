```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

image = cv2.imread('input/watermelon.png')
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
plt.imshow(image)

print(image.shape)
```

```python
R,G,B = cv2.split(image)

R_new = G
G_new = G
B_new = R

result = cv2.merge((R_new, B_new, G_new))
plt.imshow(result)
result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
cv2.imwrite('output/Convert Watermelon .jpg', result)

```
