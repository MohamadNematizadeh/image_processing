```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

image = cv2.imread('input/watermelon.png')
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
plt.imshow(image)
print(image.shape)
```

```python
R, G, B = cv2.split(img)

G_new = G
B_new = B
R_new = R

img_result = cv2.merge((G_new,R_new,B_new))

plt.imshow(img_result)
```

```python
img_result = cv2.cvtColor(img_result, cv2.COLOR_RGB2BGR)

cv2.imwrite('output1.jpg', img_result)
```
