```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("input/photo4656923068.jpg",0)
plt.imshow(img,cmap="gray")
plt.show()
```

```python
_,image_thresh = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(image_thresh,cmap="gray")
plt.show()
```

```python
contours,_=cv2.findContours(image_thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
result =img.copy()
for contour in contours:
    if cv2.contourArea(contour)> 1000:
        x,y,w,h = cv2.boundingRect(contour)
        cv2.rectangle(result,(x,y),(x+w,y+h),(255,255,255),3)

plt.imshow(result,cmap="gray")
```
