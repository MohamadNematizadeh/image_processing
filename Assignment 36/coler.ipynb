```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
img = cv2.imread("inpurt/1.jpg")
img_RGB = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img , cv2.COLOR_BGR2HSV)
H,S,V = cv2.split(img)
```

```python
H = H.copy().astype(np.float32)
H_new = H.copy().astype(np.float32) 
for i in range(H.shape[0]):
    for j in range(H.shape[1]):
        if  S[i,j] > 150 : 
            if 105 <= H[i,j] <= 105 : 
                H_new[i,j] -= 75
            if 105 < H[i,j] <= 130 : 
                H_new[i,j] -= 80
            if 0 <= H[i,j] <= 5 : 
                H_new[i,j] += 65
            if 170 <= H[i,j] <= 180 : 
                H_new[i,j] += 70

H_new = H_new.astype(np.uint8)
result_Imag_coler = cv2.merge((H_new,S,V))
result_Imag_coler  = cv2.cvtColor(result_Imag_coler, cv2.COLOR_HSV2RGB)
plt.axis("off")
plt.imshow(result_Imag_coler)
```
