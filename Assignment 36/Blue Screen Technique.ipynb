```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load and show front image
front = cv2.imread('inpurt/SuperMan.jpg')
front_rgb = cv2.cvtColor(front, cv2.COLOR_BGR2RGB)
plt.imshow(front_rgb)
plt.show()
```

```python
# load and show background image
back = cv2.imread('inpurt/sky.jpg')
back_rgb = cv2.cvtColor(back, cv2.COLOR_BGR2RGB)
plt.imshow(back_rgb)
plt.show()
```

```python
# match front and back - resizing
back = cv2.resize(back, (front.shape[1], front.shape[0]))
back_rgb = cv2.cvtColor(back, cv2.COLOR_BGR2RGB)
plt.imshow(back_rgb)
plt.show()
```

```python
front = cv2.resize(front, None, fx=0.2, fy=0.2)
back = cv2.resize(back, None, fx=0.2, fy=0.2)
front_hsv = cv2.cvtColor(front, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(front_hsv)
plt.imshow(H, cmap='gray')
```

```python
rows, cols, _ = front.shape
for i in range(rows):
    for j in range(cols):
        if 35<H[i, j]<80 and 30<S[i, j] and 80<V[i, j]:
            front[i, j] = back[i,j]


cv2.imwrite('ouput/Blue Screen Technique.png', front)
```
