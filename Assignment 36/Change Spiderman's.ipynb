```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img_Spidermans = cv2.imread('inpurt/spiderman.jpg')
img_Spidermans = cv2.cvtColor(img_Spidermans, cv2.COLOR_BGR2HSV)
H,S,V = cv2.split(img_Spidermans)
for i in range(H.shape[0]):
    for j in range(H.shape[1]):
        if H[i,j] < 15 or H[i,j] > 165 and S[i,j] > 180:
            H[i,j] += 50
        if 100 < H[i,j] < 160 and S[i,j] > 120:
            H[i,j] -= 50
result_Spidermans= cv2.merge((H, S, V))
result_Spidermans = cv2.cvtColor(result_Spidermans, cv2.COLOR_HSV2BGR)
plt.imshow(cv2.cvtColor(result_Spidermans, cv2.COLOR_BGR2RGB))
cv2.imwrite('ouput/result_Spidermans.jpg', result_Spidermans)
```
